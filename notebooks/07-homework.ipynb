Создадим небольшой тестовый датасет

```python
import pandas as pd
import warnings

warnings.filterwarnings(action='ignore')
reviews = pd.DataFrame()

ratings = [1, 2, 3, 4, 5]
texts = [
    "In the first grade, they were fed awful cutlets with pasta.",
    "It was dirty on board, the flight attendants were aggressive, they did not feed anything.",
    "They promised free meals, but in fact they only fed me awful peanuts, to which I am allergic.",
    "The staff had a poor briefing in case of an emergency.",
    "I can mention the abundance of food and helpful staff."
    
]

titles = [
    "Terrible flight",
    "Bad airline",
    "The flight is so-so", 
    "Overall, a good flight",
    "The flight was successful"
]

data = {
    'rating': ratings,
    'text': texts,
    'title': titles
}

new_reviews = pd.DataFrame(data)
reviews = pd.concat([reviews, new_reviews], ignore_index=True)
reviews.to_csv('../data/reviews.csv', index=False)
reviews
```
