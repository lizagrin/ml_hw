### Easy (+0.1)

```python
import pandas as pd

df = pd.read_csv('../data//marketing_campaign.csv', delimiter=';')
total_complains = df['Complain'].sum()
print('Количество жалоб: ', total_complains)
```

```python
from datetime import datetime
current_year = datetime.now().year # находим текущий год 
df['Age'] = current_year - df['Year_Birth']

# Фильтруем колонки, начинающиеся с Mnt 
mnt_column = df.filter(regex='Mnt', axis=1)
customers = df[(mnt_column != 0).any(axis=1)]

# Вычисляем средний возраст 
average_age_customers = customers['Age'].mean()
average_age_all = df['Age'].mean()
diff = average_age_all - average_age_customers

print('Средний возраст покупателей:', average_age_customers)
print('Средний возраст всех людей в датасете:', average_age_all)
print('Разница среднего возраста:', diff)
```

### Medium (+0.4)

```python
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations

head = {
    'Education',
    'Marital_Status',
    'Income',
    'Kidhome',
    'Teenhome',
    'Recency',
    'NumWebPurchases',
    'NumStorePurchases',
    'Age'
}

sns.set(rc={'figure.figsize':(6,3)})
for par1, par2 in combinations(head, 2):
    sns.scatterplot(x=par1, y = par2, data=df)
    plt.show()
```

### Выводы:
1. Чем выше доход у клиента, тем чаще он покупает что-то в магазине или на сайте 
2. У людей с основным образованием доход меньше, чем у людей, получивших высшее образование 
3. Люди с меньшим количеством детей имеют более высокий доход 
4. Количество покупок не зависит от возраста
5. Есть клиенты, которые совершают покупки только на сайте или только в магазине  
6. Семья с ребенком совершает меньше покупок, чем без него
